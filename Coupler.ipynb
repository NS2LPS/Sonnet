```python
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
```

```python
Coupler=rf.Network("../Sonnet_files/Directional coupler/Coupler_param1.s4p")
```

```python
freq=Coupler.frequency.f
Param=['$S_{11}$','$S_{21}$','$S_{31}$','$S_{41}$']
```

```python
Coupler.plot_s_db(0,0)
Coupler.plot_s_db(1,0)
Coupler.plot_s_db(2,0)
Coupler.plot_s_db(3,0)
Coupler.plot_s_db(2,2)
```

```python
# Access the scattering parameters (S-matrix) at a specific frequency
# S-matrix is a 3D array: (frequency, n, n)
s_matrix = Coupler.s
det= np.zeros((len(freq), 1))
```

```python
# Choose a frequency index (e.g., 0 for the first frequency point)
for freq_index in range(len(freq)):
    det[freq_index]= np.abs(np.linalg.det(s_matrix[freq_index]))
```

```python
plt.plot(1e-9*freq,np.abs(det))
plt.xlabel("frequency (GHz)")
plt.ylabel("$|det(S)|-1$")
plt.savefig('Det_Sij_coupler.svg')
plt.show()
```
